```python
# 21) What is the average Visibility_km in weekends on february and march

df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df_feb_march = df[(df['Date/Time'].dt.month == 2) | (df['Date/Time'].dt.month == 3)]
df_feb_march['Day of Week'] = df_feb_march['Date/Time'].dt.dayofweek
weekend_df = df_feb_march[(df_feb_march['Day of Week'] == 5) | (df_feb_march['Day of Week'] == 6)]
average_visibility_weekends_feb_march = weekend_df['Visibility_km'].mean()
print(average_visibility_weekends_feb_march)
```

```python
#  23) Plot the distrubution of wind speed against each 'Weather Condition'

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(x='Weather Condition', y='Wind Speed_km/h', data=df)
plt.xticks(rotation=45)
plt.title("Distribution of Wind Speed against Weather Conditions")
plt.xlabel("Weather Condition")
plt.ylabel("Wind Speed (km/h)")
plt.show()
```
